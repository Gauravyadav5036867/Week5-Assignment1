```python
#Question 3

Try and except statements are used to catch and handle exceptions in Python. Statements that can raise exceptions are kept inside the try clause and the statements that handle the exception are written inside except clause.

def divide(a, b):
    try:
        # Code that might raise an exception
        result = a / b
    except ZeroDivisionError:
        # Code to handle the exception
        print("Cannot divide by zero")
        result = None
    else:
        # Code to run if no exceptions occur
        print("Division successful")
    finally:
        # Code that will always run
        print("Execution of finally block")
    return result

# Testing the function
result = divide(10, 2)
print(f"Result: {result}")

result = divide(10, 0)
print(f"Result: {result}")
```

```python
#Question 4

Try:
    This block attempts to convert the inputs a and b to floats.
It checks if b is zero and raises a ZeroDivisionError if so.
It performs the division if all checks pass.

Else:
    This block runs only if no exceptions were raised in the try block. It prints the result of the division.
    
Finally:
    This block runs regardless of whether an exception was raised or not. It is typically used for cleanup actions or to ensure that certain code runs no matter what.
    
Raise:
    The raise statement is used to raise an exception. In this example, it raises a ZeroDivisionError if b is zero.
    
    
    
    def validate_and_divide(a, b):
    try:
        # Try to convert inputs to float
        a = float(a)
        b = float(b)
        
        # Check for division by zero
        if b == 0:
            raise ZeroDivisionError("Cannot divide by zero")
        
        result = a / b
    except ValueError:
        # Handle case where inputs are not valid numbers
        print("Error: Invalid input. Please enter numbers only.")
    except ZeroDivisionError as e:
        # Handle division by zero
        print(f"Error: {e}")
    else:
        # This block runs if no exceptions occur
        print(f"Division successful: {result}")
    finally:
        # This block always runs
        print("Execution complete")

# Test cases
validate_and_divide(10, 2)   # Valid input, should succeed
validate_and_divide(10, 0)   # Division by zero, should raise ZeroDivisionError
validate_and_divide("10", "a")  # Invalid input, should raise ValueError
```

```python
#Question 5

Custom exceptions are user-defined exceptions that allow you to create specific error conditions for your application. While Python has many built-in exceptions (like ValueError, TypeError, etc.), sometimes they may not adequately describe the error conditions in your application. Custom exceptions can provide more meaningful error messages and help in creating a more readable and maintainable codebase.

Why Do We Need Custom Exceptions:
Specific Error Handling: Custom exceptions allow you to define and handle specific error conditions that are unique to your application.
Improved Readability: They make your code more readable and self-documenting by clearly indicating what type of error has occurred.
Better Debugging: Custom exceptions can provide more context and detailed error messages, making it easier to debug and maintain the code.
Modular Code: They help in separating error handling logic from the main logic, making the code cleaner and more modular.


# Define a custom exception
class InvalidAgeError(Exception):
    """Exception raised for errors in the input age."""
    def __init__(self, age, message="Age must be between 0 and 120"):
        self.age = age
        self.message = message
        super().__init__(self.message)

    def __str__(self):
        return f'{self.age} -> {self.message}'

# Function to check age validity
def check_age(age):
    if not (0 <= age <= 120):
        raise InvalidAgeError(age)
    else:
        print(f"Age {age} is valid.")

# Testing the function with different inputs
try:
    check_age(25)  # Valid age
    check_age(-5)  # Invalid age, will raise InvalidAgeError
except InvalidAgeError as e:
    print(f"Error: {e}")

try:
    check_age(150)  # Invalid age, will raise InvalidAgeError
except InvalidAgeError as e:
    print(f"Error: {e}")
```
